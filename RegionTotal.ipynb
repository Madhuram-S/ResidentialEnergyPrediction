```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
raw_df = pd.read_csv("../Project3/dataforfinalproject/2015/recs2015_public_v4.csv")
raw_df.head()
```

```python
northeastdf = raw_df.loc[raw_df['REGIONC'] == 1]
midwestdf = raw_df.loc[raw_df['REGIONC'] == 2]
southdf = raw_df.loc[raw_df['REGIONC'] == 3]
westdf = raw_df.loc[raw_df['REGIONC'] == 4]


westdf.head(20)
```

```python
northeast = northeastdf[['REGIONC', 'DIVISION', 'HDD30YR', 'CDD30YR', 'KWH', 'TOTALBTU', 'KWHAHUHEAT', 'KWHAHUCOL']].groupby(['REGIONC', 'DIVISION'])
midwest = midwestdf[['REGIONC', 'DIVISION', 'HDD30YR', 'CDD30YR', 'KWH', 'TOTALBTU', 'KWHAHUHEAT', 'KWHAHUCOL']].groupby(['REGIONC', 'DIVISION'])
south = southdf[['REGIONC', 'DIVISION', 'HDD30YR', 'CDD30YR', 'KWH', 'TOTALBTU', 'KWHAHUHEAT', 'KWHAHUCOL']].groupby(['REGIONC', 'DIVISION'])
west = westdf[['REGIONC', 'DIVISION', 'HDD30YR', 'CDD30YR', 'KWH', 'TOTALBTU', 'KWHAHUHEAT', 'KWHAHUCOL']].groupby(['REGIONC', 'DIVISION'])


northeast.head()
```

```python
#find the difference between HDD and CDD for each region

northeastsum = northeast.sum()
midwestsum = midwest.sum()
southsum = south.sum()
westsum = west.sum()

northeastsum.head()
```
